## Titanic data analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("titanic/train.csv")
test_data = pd.read_csv('titanic/test.csv')

print(data.shape)
data.isna().sum()
```

```python
data.info()
```

## Description of columns


```python
# Check for NA values

print(F"In training data { data.isna().sum()}") 
print(F"\n In testing data {test_data.isna().sum() }")

# Age column has 177 Null values
```

```python
data.describe()
```

### Distribution Of Age

```python
# Let's replace the missing values of Age  

print("Distribution of Age")
data["Age"].hist(bins = 20)   # We have right skewed distribution
print(F'Median = {data["Age"].median()}. Mean = {data["Age"].mean()}')
```

We analyze the disribution of age in different classes. Because median age differs in passenger classes. Professional advancement usually comes with increasing age and experience. Also we need to consider the gender.  

```python
# Distribution of age across passenger classes
print("Distribution of age across passenger classes")
print(F"Median of ages in different passenger classes = \
            {data.groupby( ['Pclass'] )['Age'].median()}")


# This is super useful and saves lot of work 
data["Age"].hist(by = data["Pclass"])    # look at the docs of df.groupby(by = ). We can group them in multiple ways


```

```python
# Distribution of age by passenger class and then by gender

print("Distribution of age in each gender in each classes")
print(F"Median of ages og each gender in each passenger class = \
            { data.groupby(['Pclass', 'Sex'])['Age'].median() }")

# This is not a very good way of plotting i will improve this in future

fig, ax = plt.subplots(3,2, figsize = (20, 15))
fig.subplots_adjust(hspace = 0.4, wspace = 0.2)

i = 0
j = 0

for grp in data.groupby(["Pclass", "Sex"])["Age"]:
    
    # grp[0] - (1, female)
    # grp[1] - Series object containing all ages 
    
    if j == 2:
        j = 0 
        i += 1
    ax[i][j].hist(grp[1].dropna().values)
    ax[i][j].set_title( grp[0][1] + " class {}".format(grp[0][0]), fontsize = 15)
    j += 1
    
    
```

We can observe that in higher classes people are more aged. Also in a specific class, men are more aged than women. 

So we replace each missing value of age by median in their passenger class and gender 

```python
data["Age"] = data.groupby(["Pclass", "Sex"])["Age"].apply(lambda x : x.fillna(x.median()))  
# x is a Series object here    

test_data["Age"] = test_data.groupby(["Pclass", "Sex"])["Age"].apply(lambda x : x.fillna(x.median()))  
```

```python
print(data.isna().sum())
print(test_data.isna().sum())
```

Embarked has 2 missing values. On some googling we can find that both of the passengers embarked on "S"

```python
print(data[data["Embarked"].isna()]) 

data["Embarked"].fillna("S", inplace = True)

data.isna().sum()
```

There are lot of missing values in cabin column. But we can take the first variable of cabin number which is A, B, C, D, .., F. A is located at topmost and F is located bottommost. Let's add another column "Deck" where we store only first letter of cabin number if availible, otherwise "M" (indicating missing)   

```python
data["Deck"] = data["Cabin"].apply(lambda x : x[0] if pd.notnull(x) else "M")
data["Deck"].value_counts()
```

## Feature Engineering
- Binning continuous variable ("Age"). 
- Creating new features out of existing features ("Title").
- One hot Encoding for nominal varaibles ("Sex").
- Label Encoding for ordinal variables ("Pclass"). 

### Outlier detection

```python
data.boxplot(column = ["Fare"], figsize = (5,7))
```

```python
data.boxplot(column = ["Age"], figsize = (5,7))
```

We can see that both Fare and Age columns have outliers. 

```python
data["Fare_binned"] = pd.qcut(data["Fare"], 5)  # divide into 5 equal quantiles
data["Age_binned"] = pd.cut(data["Age"].astype(int) , 5)  # divide into 5 equal bins - int just to round off
```

We expect young people and people who have paid high fare to have high survival rate. Young people would be probably rescued first and people with high ticket prices could be gicen lifeboats first

```python
print("Count of Fare in each bin")
print(data["Fare_binned"].value_counts().sort_index())

print("\n Count of Age in each bin")
print(data["Age_binned"].value_counts().sort_index())

print("\n Mean number of people who survived in each age category and each fare category")
print(data[["Age_binned", "Survived"]].groupby("Age_binned")["Survived"].mean(), "\n")
print(data[["Fare_binned", "Survived"]].groupby("Fare_binned")["Survived"].mean())
```

We can see that our assumption turned out to be true. Most of the saved people are younger and most of the who paid high fare people has survived. 

### Bar chart depicting the above result

```python
data[["Age_binned", "Survived"]].groupby("Age_binned")["Survived"].mean().plot(kind = 'bar', figsize = (10, 5))   
plt.suptitle("Survival rates for age categories")
plt.xticks(rotation = 45, ha = "right")
```

```python
data[["Fare_binned", "Survived"]].groupby("Fare_binned")["Survived"].mean().plot(kind = 'bar', figsize = (10, 5))   
plt.suptitle("Survival rates for fare categories")
plt.xticks(rotation = 45, ha = "right")
```

## Family size
There are two interesting variables in our data set which tells us something about family size. SibSp defines how many siblings and spouses a passenger had and parch how many parents and childrens. We can summarize these variables and add 1 (for each passer-by) to get the family size.


```python
data["Family_size"] = data["SibSp"] + data["Parch"] + 1
test_data["Family_size"] = test_data["SibSp"] + test_data["Parch"] + 1
data["Family_size"].hist()
plt.xticks(rotation = 45, ha = "right")
```

We expect that moderate sized families have high chance of survival because of mutual support. But this could be difficult if families are large sized. 

```python
data[["Family_size", "Survived"]].groupby(["Family_size"])["Survived"].mean().plot(kind = "bar")  
plt.xticks(rotation = 0, ha = "right")
```

As we expected families of size 4 have hughest survival rate. 

## Title 
The name od the person has title (Mr, Miss, Mrs). We can answer the question of if someone is married or not if someone has formal title it could be an indicator of higher sociol status.  

```python
print(data["Name"].str.split(',', expand = True)[1].str.split('.', expand = True)[0].str.strip().head())
data["Title"] = data["Name"].str.split(',', expand = True)[1].str.split('.', expand = True)[0].str.strip()
test_data["Title"] = test_data["Name"].str.split(',', expand = True)[1].str.split('.', expand = True)[0].str.strip()

print("\n Counts of people's of each titles")
print(data["Title"].value_counts())

# Since some titles have very less people, we want to keep only those titles with more than 10 people 
# and assign Misc to others

titles_lt10 = (data["Title"].value_counts() < 10) 
data["Title"] = data["Title"].apply(lambda x : "Misc" if titles_lt10.loc[x] else x)
print("\n After filtering titles > 10 members")
print(data["Title"].value_counts())

# Same thing on testing data 

titles_lt10 = (test_data["Title"].value_counts() < 10) 
test_data["Title"] = test_data["Title"].apply(lambda x : "Misc" if titles_lt10.loc[x] else x)

# We can see that there are lot of titles but we are interested in only few important ones
print("\n Survival rate of people wrt title")
print(data[["Title", "Survived", "Family_size"]].groupby("Title").mean().sort_values(by = "Survived", ascending = False) )
```

```python
print(data.columns)
print(test_data.columns)
```

Missing values in test data 

```python
print(test_data.isna().sum())

# There is one missing value in Fare, we will replace it with median value of Fare in that passenger class

pclass_na = test_data.loc[test_data["Fare"].isna(), "Pclass"].values[0]
fare_na = test_data.loc[test_data["Pclass"] == pclass_na, "Fare"].median()
test_data["Fare"] = test_data["Fare"].fillna(fare_na)
test_data["Fare"].isna().sum()
```

We will remove some columns from data, as we added them just for analysis 

```python
data.drop(columns = ["Age_binned", "Fare_binned", "Name", "Ticket", "Cabin"] , inplace = True)
test_data.drop(columns = ["Name", "Ticket", "Cabin"], inplace = True)

print(data.columns)
print(test_data.columns)
```

Let's encode categorical columns 

```python
# Title column

title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Misc": 5}

data["Title"] = data["Title"].map(title_mapping)
test_data["Title"] = test_data["Title"].map(title_mapping)

print(data["Title"].value_counts())
print(test_data["Title"].value_counts())
```

```python
# Gender column 

data["Sex"] = data["Sex"].map({"female" : 1, "male" : 0})
test_data["Sex"] = test_data["Sex"].map({"female" : 1, "male" : 0})

print(data["Sex"].value_counts())
print(test_data["Sex"].value_counts())
```

```python
# Age 
# Based on the Age_binned column we used above we can convert Age into a ordinal categorical column

combine = [data, test_data]

for dataset in combine:    
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age'] = 4

data["Age"].value_counts()
```

We can create a column isAlone and drop SibSp, Parch, Family_size columns

```python
combine = [data, test_data]

for dataset in combine:
    dataset["IsAlone"] = 0
    dataset.loc[dataset["Family_size"] == 1, "IsAlone"] = 1

print(data[["IsAlone", "Survived"]].groupby("IsAlone", as_index = False).mean())

# When as_index=True the key(s) you use in groupby() will become an index in the new dataframe.

data.drop(columns = ["SibSp","Parch","Family_size"], inplace = True)
test_data.drop(columns = ["SibSp","Parch","Family_size"], inplace = True)
```
